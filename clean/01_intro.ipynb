There are 7,393 items, but all of them may not be images. The `get_image_files()` function returns the images in a folder.

```python
doc(ImageDataLoaders.from_name_func)
```

```python
dls = ImageDataLoaders.from_name_func(
    path=path,
    fnames=get_image_files(path),
    label_func=is_cat,
    valid_pct=0.2,
    seed=42,
    item_tfms=Resize(224),
    shuffle=True
)
type(dls)
```

**Note:** The `seed` argument is only used for splitting the dataset into a training & a validation set. Hence, there will still be some random variation in the results (if you re-run the training multiple times). You can prevent this by seeding everything.

```python
Resize
```

**Note:** `Resize` is probably equivalent to `torchvision.transforms.Resize`.

```python
doc(vision_learner)
```

```python
learn = vision_learner(dls, resnet34, normalize=True, pretrained=True, metrics=error_rate)
type(learn)
```

**Note:** Specifying `resnet34` is the same as specifying `torchvision.models.resnet34(pretrained=True)`.

```python
doc(learn.fine_tune)
```

**Note:** Some of the parameters overlap with `vision_learner()`, such as `wd`, `moms` and `cbs`.

```python
learn.fine_tune(epochs=1, freeze_epochs=1)
```

```python
doc(TextDataLoaders.from_folder)
```

**Note:**

1. The factory method `TextDataLoaders.from_folder` is suitable when (i) you want to build a text classifier (`is_lm=False`) or a language model (`is_lm=True`) and (ii) the data has an ImageNet style folder structure (as in this example). For other modeling scenarios (e.g., regression), use other factory methods / the DataBlock API / the mid-level API.
2. The `text_vocab` argument can be used to update the default text vocabulary.
3. The `seq_len` argument can be used to set the sequence length.
4. The `backwards` argument can be used to specify that we'll be using a backwards model, and to prepare the data accordingly.

```python
dls = TextDataLoaders.from_folder(path=path, valid='test', shuffle=True)
type(dls)
```

**Question:** How does `TextDataLoaders.from_folder` know which tokenizer to use? (We haven't specified a checkpoint yet; we'll do so in `text_classifier_learner`.) Also, does the fastai library contain only one pretrained model (`AWD_LSTM`)?

`dls.vocab[1]` contains the class names:

```python
dls.vocab[1]
```

`TextDataLoaders.from_folder` inferred the class names from the sub-folder names inside `'train'` and `'test'`.

```python
doc(text_classifier_learner)
```

**Note:** It seems that `seq_len` and `backwards` (if changed from their default values) need to be specified in both `TextDataLoaders.from_folder` and `text_classifier_learner`.

```python
learn = text_classifier_learner(dls, AWD_LSTM, drop_mult=0.5, metrics=accuracy)
type(learn)
```

```python
learn.fine_tune(epochs=4, base_lr=1e-2, freeze_epochs=1)
```
